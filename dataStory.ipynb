```
import csv

with open('savings_data_final.csv') as f:
  reader = csv.reader(f)
  savingsData = list(reader)

savingsData.pop(0)
```

Finding total number of people and number of people who were reminded.

```
totalEntries = len(savingsData)
totalPeopleGivenReminder = 0

for data in savingsData:
  if int(data[3]) == 1:
    totalPeopleGivenReminder += 1

import plotly.graph_objects as go 
fig = go.Figure(go.Bar(x=['Reminded People', 'Not Remineded People'], y=[totalPeopleGivenReminder, totalEntries]))
fig.show()
```

Mean, median & mode


```
allSavings = []
for data in savingsData:
  allSavings.append(float(data[0]))

print(statistics.mean(allSavings))
print(statistics.mode(allSavings))
print(statistics.median(allSavings))
```

```
listOfRemindedSavings = []
listOfNotRemindedSavings = []

for data in savingsData:
  if int(data[3]) == 1:
    listOfRemindedSavings.append(data[0])
  else:
    listOfNotRemindedSavings.append(data[0])

print(listOfRemindedSavings)
print(listOfNotRemindedSavings)
```
